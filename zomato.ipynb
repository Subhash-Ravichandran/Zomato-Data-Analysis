## **Creating new column on Currency conversion in INR**

```python
df=df.insert(0, "Amount_in_INR", pd.NA)
```

### **Replacing with correct spellings**

```python
#哿stanbul,S恚o Paulo,Bras韄lia
# Define a dictionary with incorrect and correct spellings
replacements = {"哿stanbul": "Istanbul","S恚o Paulo": "Sao Paulo","Bras韄lia": "Brasília"}
df['City'] = df['City'].replace(replacements)

```

## **Function to convert Currencies into INR**

```python
# Currency conversion dictionary
C_Convert = {1: 1, 14: 55.52, 30: 16.24, 37: 61.13, 94: 0.0052, 148: 50.50, 162: 1.44, 166: 22.89,184: 61.88, 189: 4.56, 191: 0.28, 208: 2.58, 214: 22.70, 215: 105.53, 216: 83.36}

# Define the currency conversion function
def currency_convert(row):
    average_cost = row['Average_Cost_for_two']
    country_code = row['Country_Code']
    if country_code in C_Convert:
        return average_cost * C_Convert[country_code]
    else:
        return None

# Apply the function to each row in the DataFrame
df['Amount_in_INR'] = df.apply(currency_convert, axis=1)
```

## **Handling Cuisines**

```python
cuisines = df['Cuisines'].value_counts(ascending=False)
cuisines_less_than_100 = cuisines[cuisines < 100]
def replace_cuisines(value):
    if (value in cuisines_less_than_100):
        return 'Other'
    else:
        return value


df['Cuisines'] = df["Cuisines"].apply(replace_cuisines)
df["Cuisines"].value_counts()
```
